```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read in tables
geography = pd.read_csv('Covid_Data/geography.csv', encoding='latin-1')
geography_industry = pd.read_csv('Covid_Data/geography_industry.csv', encoding='latin-1')
industry = pd.read_csv('Covid_Data/industry.csv')
industry_job_family = pd.read_csv('Covid_Data/industry_job_family.csv')
job_family = pd.read_csv('Covid_Data/job_family.csv')
job_family_role = pd.read_csv('Covid_Data/job_family_role.csv')
overall = pd.read_csv('Covid_Data/overall.csv')
```

Here I made tables for growth/decline for each industry for 1 month, 3 months, 6 months, 9 months, 12 months.

```python
# convert to usable time format, transform 

industry['post_date'] = industry.post_date.apply(pd.Timestamp)
industry['Percent (%) Change'] = (industry['count_id_indexed'] - 1) * 100
df = industry[(industry['post_date'].dt.month == 3) & (industry['post_date'].dt.year == 2020)]
```

```python
# 1 month

grouped_1 = df.groupby('industry').mean()[['Percent (%) Change']].sort_values(by='Percent (%) Change')
grouped_1 = grouped_1.rename_axis('Industry')
f = {'Percent (%) Change':'{:.2f}'}
(
    grouped_1.style
    .set_table_attributes("style='display:inline'")
    .set_caption('Percent Change by Industry in March 2020')
    .format(f)
)
```

```python
# 3 month

df_2 = industry[(industry['post_date'].dt.month <= 5) & (industry['post_date'].dt.year == 2020)]
grouped_2 = df_2.groupby('industry').mean()[['Percent (%) Change']].sort_values(by='Percent (%) Change')
grouped_2 = grouped_2.rename_axis('Industry')
(
    grouped_2.style
    .set_table_attributes("style='display:inline'")
    .set_caption('Percent Change by Industry from March to May 2020 (3 months)')
    .format(f)
)
```

```python
# 6 month

df_3 = industry[(industry['post_date'].dt.month <= 8) & (industry['post_date'].dt.year == 2020)]
grouped_3 = df_3.groupby('industry').mean()[['Percent (%) Change']].sort_values(by='Percent (%) Change')
grouped_3 = grouped_3.rename_axis('Industry')
(
    grouped_3.style
    .set_table_attributes("style='display:inline'")
    .set_caption('Percent Change by Industry from March to August 2020 (6 months)')
    .format(f)
)
```

```python
# 9 month

df_4 = industry[(industry['post_date'].dt.month <= 12) & (industry['post_date'].dt.year == 2020)]
grouped_4 = df_4.groupby('industry').mean()[['Percent (%) Change']].sort_values(by='Percent (%) Change')
grouped_4 = grouped_4.rename_axis('Industry')
(
    grouped_4.style
    .set_table_attributes("style='display:inline'")
    .set_caption('Percent Change by Industry from March to December 2020 (9 months)')
    .format(f)
)
```

```python
# 1 year

df_5 = industry[((industry['post_date'].dt.year == 2020)) | (industry['post_date'].dt.month <= 2)]
grouped_5 = df_5.groupby('industry').mean()[['Percent (%) Change']].sort_values(by='Percent (%) Change')
grouped_5 = grouped_5.rename_axis('Industry')
(
    grouped_5.style
    .set_table_attributes("style='display:inline'")
    .set_caption('Percent Change by Industry from March 2020 to Feb 2021 (1 year)')
    .format(f)
)
```
